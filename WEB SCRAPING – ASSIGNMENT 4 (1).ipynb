```python
driver=webdriver.Chrome()
driver.get('https://archive.ics.uci.edu/')
```

```python
search_button=driver.find_element(By.XPATH,'//div[@class="flex flex-wrap justify-center gap-5"]/a[1]')
search_button.click()
```

```python
search_button2=driver.find_element(By.XPATH,'//span[@class="swap-on text-primary-content"]')
search_button2.click()
```

```python
dataset_name=driver.find_elements(By.XPATH, '//a[@class="link-hover link text-xl font-semibold"]')
dataset_type=driver.find_elements(By.XPATH, '//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[2]/span')
task=driver.find_elements(By.XPATH, '//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[1]/span')
attribute_type=driver.find_elements(By.XPATH, '//table[@class="col-span-full my-2 table sm:col-start-2"]/tbody/tr/td[2]')
no_instance=driver.find_elements(By.XPATH, '//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[3]/span')
no_attribute=driver.find_elements(By.XPATH, '//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[4]/span')
year=driver.find_elements(By.XPATH, '//table[@class="col-span-full my-2 table sm:col-start-2"]/tbody/tr/td[3]')
```

```python
dataset_names=[]
dataset_types=[]
tasks=[]
attribute_types=[]
no_instances=[]
no_attributes=[]
years=[]
for i in range(len(dataset_name)):
    dataset_names.append(dataset_name[i].text)
    dataset_types.append(dataset_type[i].text)
    tasks.append(task[i].text)
    attribute_types.append(attribute_type[i].text)
    no_instances.append(no_instance[i].text)
    no_attributes.append(no_attribute[i].text)
    years.append(year[i].text)
```

```python
df8=pd.DataFrame({'Dataset Name':dataset_names,'Dataset Type':dataset_types, 'Tasks':tasks, 'Attribute Type':attribute_types,'No of Instance':no_instances, 'No of Attributes':no_attributes, 'Dates':years})

df8
```
